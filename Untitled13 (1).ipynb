```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

data=pd.read_csv("displacement_data_1.csv")

# Sample data
x =data[(data['x']>0.01) & (data['y']>0.63) & (data['y']<0.66)]['x']
y =data[(data['x']>0.01) & (data['y']>0.63) & (data['y']<0.66)]['u_x']

x = x.values.reshape(-1, 1)

# Create polynomial features
polynomial_features = PolynomialFeatures(degree=3)
x_poly = polynomial_features.fit_transform(x)

# Create a linear regression model
model = LinearRegression()
model.fit(x_poly, y)

# Predict values
y_pred = model.predict(x_poly)

# Print the coefficients
print("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)

# Visualize the data and the fitted curve
import matplotlib.pyplot as plt

plt.scatter(x, y, color='blue')
plt.scatter(x, y_pred, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Regression')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def double_gaussian(x, A1, mu1, sigma1, A2, mu2, sigma2):
    return A1 * np.exp(-(x-mu1)**2 / (2*sigma1**2)) + A2 * np.exp(-(x-mu2)**2 / (2*sigma2**2))

# Initial guess for parameters
p0 = [10**-6, 0.25, 0.083, 10**-6, 0.75, 0.083]

# Fit the curve
popt, pcov = curve_fit(double_gaussian, x_val, slope_val, p0=p0)

# Plot the data and the fitted curve
plt.plot(x_val, slope_val, 'bo', label='Data')
plt.plot(x_val, double_gaussian(x_val, *popt), 'r-', label='Fitted Curve')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
