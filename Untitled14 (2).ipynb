```python
def sigma_yy(x,y):
  if (x>0.7 and x<0.3) and (y>0.6 or y<0.4):
      return (323.78e9 * (9.5e-5 * (1 - np.exp(-25.5 * (x - 0.5)**2)))) + \
         (147.38e9 * (3.6e-5 - (9.4e-6 * (np.exp(-50 * (y - 0.3)**2) + np.exp(-50 * (y - 0.7)**2)))))
  elif 0.3 < x < 0.7 and (y<0.4 or y>0.6):
      return (323.78e9 * (9.5e-5 * (1 - np.exp(-25.5 * (x - 0.5)**2)))) + \
         (147.38e9 * (3.6e-5 - (9.4e-6 * (np.exp(-50 * (y - 0.3)**2) + np.exp(-50 * (y - 0.7)**2))) + y * (1 - y) * (0.85 * 10**-2 * (x - x**2 - 0.21))))
  elif 0.4 < y < 0.6 and (x < 0.3 or x > 0.7):
    return (323.78e9 * (9.5e-5 * (1 - np.exp(-25.5 * (x - 0.5)**2)))) + \
           (147.38e9 * (3.6e-5 - (9.4e-6 * (np.exp(-50 * (y - 0.3)**2) + np.exp(-50 * (y - 0.7)**2))) + \
           (x * (1 - x) * (1.33e-2 * (y - y**2 - 0.24)))))
  else:
    return (323.78e9 * (9.5e-5 * (1 - np.exp(-25.5 * (x - 0.5)**2)) + \
           (y * (1 - y) * (1.85e2 * (x - x**2 - 0.21))))) + \
           (147.38e9 * (3.6e-5 - (9.4e-6 * (np.exp(-50 * (y - 0.3)**2) + np.exp(-50 * (y - 0.7)**2))) + \
           (x * (1 - x) * (1.33e-2 * (y - y**2 - 0.24)))))

```

```python
def sigma_xx(x,y):
  if (x < 0.3 or x > 0.7) and (y < 0.4 or y > 0.6):
    return (323.78e9 * (3.6e-5 - (9.4e-6 * (np.exp(-50 * (y - 0.3)**2) + np.exp(-50 * (y - 0.7)**2))))) + \
           (147.38e9 * (9.5e-5 * (1 - np.exp(-25.5 * (x - 0.5)**2))))
  elif (x < 0.3 or x > 0.7) and 0.4 < y < 0.6:
    return (323.78e9 * (3.6e-5 - (9.4e-6 * (np.exp(-50 * (y - 0.3)**2) + np.exp(-50 * (y - 0.7)**2))))) + \
           (147.38e9 * (9.5e-5 * (1 - np.exp(-25.5 * (x - 0.5)**2)) + \
           (x * (1 - x) * (1.33e-2 * (y - y**2 - 0.24)))))
  elif (y<0.4 or y>0.6) and (0.3<x<0.7):
    return (323.78e9 * (3.6e-5 - (9.4e-6 * (np.exp(-50 * (y - 0.3)**2) + np.exp(-50 * (y - 0.7)**2)))+(y * (1 - y) * (0.185e-2 * (x - x**2 - 0.21))))) + \
        (147.38e9 * (9.5e-5 * (1 - np.exp(-25.5 * (x - 0.5)**2))))
  else:
    return (323.78e9 * (3.6e-5 - (9.4e-6 * (np.exp(-50 * (y - 0.3)**2) + np.exp(-50 * (y - 0.7)**2)))+(y * (1 - y) * (0.185e-2 * (x - x**2 - 0.21))))) + \
            (147.38e9 * (9.5e-5 * (1 - np.exp(-25.5 * (x - 0.5)**2)) + \
            (x * (1 - x) * (1.33e-2 * (y - y**2 - 0.24)))))
```

```python
def tau_xy(x,y):
  if x<0.3 and y<0.4:
      return (4.845e-3 * y * (x - 0.5) * np.exp(-25.5 * (x - 0.5)**2)) + \
         (9.4e-4 * x * ((y - 0.3) * np.exp(-50 * (y - 0.3)**2) + (y - 0.7) * np.exp(-50 * (y - 0.7)**2)))
  elif x>0.7 and y<0.4:
      return (4.845e-3 * y * (x - 0.5) * np.exp(-25.5 * (x - 0.5)**2)) + \
         (9.4e-4 * x * (1 - x) * ((y - 0.3) * np.exp(-50 * (y - 0.3)**2) + (y - 0.7) * np.exp(-50 * (y - 0.7)**2)))
  elif 0.3<x<0.7 and y<0.4:
      return (4.845e-3 * y * (x - 0.5) * np.exp(-25.5 * (x - 0.5)**2)) + \
         (9.4e-4 * x * ((y - 0.3) * np.exp(-50 * (y - 0.3)**2) + (y - 0.7) * np.exp(-50 * (y - 0.7)**2))) + \
         (0.185e-2 * (x / 2 - x**2 / 3 - 0.21) + 6.8e-5) * (1 - 2*y)
  elif y>0.6 and x<0.3:
      return (4.845e-3 * (y - 1) * (x - 0.5) * np.exp(-25.5 * (x - 0.5)**2)) + \
         (9.4e-4 * x * ((y - 0.3) * np.exp(-50 * (y - 0.3)**2) + (y - 0.7) * np.exp(-50 * (y - 0.7)**2)))
  elif y>0.6 and x>0.7:
      return (4.845e-3 * (y - 1) * (x - 0.5) * np.exp(-25.5 * (x - 0.5)**2)) + \
         (9.4e-4 * (x - 1) * ((y - 0.3) * np.exp(-50 * (y - 0.3)**2) + (y - 0.7) * np.exp(-50 * (y - 0.7)**2)))
  elif 0.3<x<0.7 and y>0.6:
      return (4.845e-3 * (y - 1) * (x - 0.5) * np.exp(-25.5 * (x - 0.5)**2)) + \
         (9.4e-4 * (x - 1) * ((y - 0.3) * np.exp(-50 * (y - 0.3)**2) + (y - 0.7) * np.exp(-50 * (y - 0.7)**2))) + \
         (0.185e-2 * (x / 2 - x**2 / 3 - 0.21) + 6.8e-5) * (1 - 2*y)
  elif x<0.3 and 0.4<y<0.6:
      return (4.845e-3 * y * (x - 0.5) * np.exp(-25.5 * (x - 0.5)**2)) + \
         (1.33e-2 * y * (y / 2 - y**2 / 3 - 0.24) + 5e-5) * (1 - 2*x) + \
         (9.4e-4 * x * ((y - 0.3) * np.exp(-50 * (y - 0.3)**2) + (y - 0.7) * np.exp(-50 * (y - 0.7)**2)))
  elif x>0.7 and 0.4<y<0.6:
      return (4.845e-3 * y * (x - 0.5) * np.exp(-25.5 * (x - 0.5)**2)) + \
         (1.33e-2 * y * (y / 2 - y**2 / 3 - 0.24) + 5e-5) * (1 - 2*x) + \
         (9.4e-4 * x * ((y - 0.3) * np.exp(-50 * (y - 0.3)**2) + (y - 0.7) * np.exp(-50 * (y - 0.7)**2)))
  else:
      return (4.845e-3 * y * (x - 0.5) * np.exp(-25.5 * (x - 0.5)**2)) + \
         (1.33e-2 * y * (y / 2 - y**2 / 3 - 0.24) + 5e-5) * (1 - 2*x) + \
         (9.4e-4 * x * ((y - 0.3) * np.exp(-50 * (y - 0.3)**2) + (y - 0.7) * np.exp(-50 * (y - 0.7)**2))) + \
         (0.185e-2 * (x / 2 - x**2 / 3 - 0.21) + 6.8e-5) * (1 - 2*y)
```

```python
sxx=[]
syy=[]
txy=[]
for i in range(len(data['x'])):
  sxx.append(sigma_xx(data['x'][i],data['y'][i]))
  syy.append(sigma_yy(data['x'][i],data['y'][i]))
  txy.append(tau_xy(data['x'][i],data['y'][i]))


```

```python
data['sigma_xx']=sxx
data['sigma_yy']=syy
data['tau_xy']=txy
```

```python
x1=np.linspace(0,1,700)
y1=np.linspace(0,1,700)
stress_val=[]
for i in range(len(y1)):
  stress_val1=[]
  for j in range(len(x1)):
    stress_val1.append(sigma_xx(x1[i],y1[j]))
  stress_val.append(stress_val1)
stress_val=np.array(stress_val)
X1,Y1=np.meshgrid(x1,y1)
```
